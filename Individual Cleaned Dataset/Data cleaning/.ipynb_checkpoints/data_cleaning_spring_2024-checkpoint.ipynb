```python
import pandas as pd
import numpy as np

df = pd.read_csv('THM_Spring_2024.csv')

print(f"Total rows: {len(df)}")  # Should show 91
print(f"Total columns: {len(df.columns)}")  # Should show 56
```

```python
# Shorter, cleaner column names
df.columns = [
    # Metadata
    'StartDate', 'EndDate', 'Status', 'IPAddress', 'Progress', 'Duration',
    'Finished', 'RecordedDate', 'ResponseID', 'RecipientLastName',
    'RecipientFirstName', 'RecipientEmail', 'ExternalReference',
    'LocationLatitude', 'LocationLongitude', 'DistributionChannel', 
    'UserLanguage',
    
    # Consent
    'ConsentText',
    
    # IDs
    'StudentID_1', 'StudentID_2',
    
    # Demographics
    'Demo_HispanicLatinx', 'Demo_Race', 'Demo_Gender', 'Demo_ParentEducation',
    
    # Policy Area
    'PolicyArea',
    
    # Civic Engagement Fair
    'CivicFair_Orgs',
    'CivicFair_OrgsMissed',
    'CivicFair_Learning',
    'CivicFair_Learning_Other',
    'CivicFair_PolicyConnection',     # Q6_1
    'CivicFair_CampusUnderstanding',  # Q6_2
    'CivicFair_UsedInBreakout',       # Q6_3
    'CivicFair_PreparedForBreakout',  # Q6_4
    
    # Breakout Rooms
    'Breakout_Learning',
    'Breakout_Learning_Other',
    'Breakout_POLSPreparation',       # Q9_1
    'Breakout_Participation',         # Q9_2
    'Breakout_ConsultantContribution', # Q9_3
    'Breakout_DiscussionBalance',     # Q9_4
    'Breakout_CivicFairConnection',   # Q9_5
    
    # Action Symposium
    'Action_StationsChosen',
    'Action_StationsWanted',
    'Action_Learning',
    'Action_Learning_Other',
    'Action_TalkToExperts',           # Q14_1
    'Action_ConsultantKnowledge',     # Q14_2
    'Action_CivicFairConnection',     # Q14_3
    'Action_BreakoutConnection',      # Q14_4
    
    # Overall Impact (Before/After pairs)
    'Overall_Connection_Before',
    'Overall_Connection_After',
    'Overall_AcademicConf_Before',
    'Overall_AcademicConf_After',
    'Overall_ChangeConf_Before',
    'Overall_ChangeConf_After',
    
    # Recommendation & Feedback
    'Recommend',
    'Overall_Feedback'
]

print("Columns renamed successfully!")
print(f"New columns: {df.columns.tolist()}")
```

```python
drop_cols = [
    'ConsentText',           # Long consent text
    'RecipientLastName',     # 100% missing
    'RecipientFirstName',    # 100% missing
    'RecipientEmail',        # 100% missing
    'ExternalReference',     # 100% missing
    'ResponseID',            # Qualtrics ID
    'IPAddress',             # Privacy
    'DistributionChannel',   # All same
    'UserLanguage'           # All EN
]

df = df.drop(columns=drop_cols)

print(f"Columns remaining: {len(df.columns)}")
```

```python
# Check if both ID columns match
df['ID_Match'] = df['StudentID_1'] == df['StudentID_2']

# Statistics
print(f"Total responses: {len(df)}")
print(f"IDs don't match: {(~df['ID_Match']).sum()}")
print(f"Missing IDs: {df['StudentID_1'].isna().sum()}")

# Create single ID column
df['StudentID'] = df['StudentID_1']

# Clean: remove mismatched or missing
print(f"\nBefore ID cleaning: {len(df)} rows")
df = df[df['ID_Match'] == True].copy()
df = df.dropna(subset=['StudentID'])
print(f"After ID cleaning: {len(df)} rows")

# Drop redundant
df = df.drop(columns=['StudentID_1', 'StudentID_2', 'ID_Match'])
```

```python
df['StartDate'] = pd.to_datetime(df['StartDate'], format='%m/%d/%Y %H:%M')
df['EndDate'] = pd.to_datetime(df['EndDate'], format='%m/%d/%Y %H:%M')
df['RecordedDate'] = pd.to_datetime(df['RecordedDate'], format='%m/%d/%Y %H:%M')

# B. Already numeric (Progress, Duration already int64)
# Just verify
print("Progress type:", df['Progress'].dtype)
print("Duration type:", df['Duration'].dtype)

# C. Location as float (already correct)

# D. StudentID as string
df['StudentID'] = df['StudentID'].astype(str).str.replace('.0', '', regex=False)

print("Data types converted!")
```

```python
print("Progress distribution:")
print(df['Progress'].value_counts())

print("\nFinished distribution:")
print(df['Finished'].value_counts())

print(f"\nBefore filtering: {len(df)} rows")

df = df[
    (df['Progress'] == 100) &
    (df['Finished'] == True)
].copy()

print(f"After filtering: {len(df)} rows")
```

```python
# Duration statistics
print("Duration (seconds) statistics:")
print(df['Duration'].describe())

# Visualize
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df['Duration'], bins=30, edgecolor='black')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.title('Survey Completion Time Distribution')
plt.axvline(120, color='red', linestyle='--', label='2 min threshold')
plt.legend()
plt.show()

# Check how many are too fast
print(f"\nResponses < 60s: {(df['Duration'] < 60).sum()}")
print(f"Responses < 120s: {(df['Duration'] < 120).sum()}")
print(f"Responses < 300s (5min): {(df['Duration'] < 300).sum()}")

# Filter (suggest 2 minutes minimum for quality)
threshold = 120  # Adjust as needed
print(f"\nBefore duration filter: {len(df)} rows")
df = df[df['Duration'] >= threshold].copy()
print(f"After duration filter: {len(df)} rows")
```

```python
df['Demo_HispanicLatinx'] = df['Demo_HispanicLatinx'].str.strip()

print("Hispanic/Latinx Status:")
print(df['Demo_HispanicLatinx'].value_counts())
print(f"Missing: {df['Demo_HispanicLatinx'].isna().sum()}")
```

```python
df['Demo_Race'] = df['Demo_Race'].str.strip()

print("\nRace/Ethnicity:")
print(df['Demo_Race'].value_counts())
print(f"Missing: {df['Demo_Race'].isna().sum()}")

# Note: Some may have multiple races listed
# Check for commas
has_multiple = df['Demo_Race'].str.contains(',', na=False).sum()
print(f"Students with multiple races: {has_multiple}")
```

```python
df['Demo_Gender'] = df['Demo_Gender'].str.strip()

print("\nGender Identity:")
print(df['Demo_Gender'].value_counts())
print(f"Missing: {df['Demo_Gender'].isna().sum()}")
```

```python
df['Demo_ParentEducation'] = df['Demo_ParentEducation'].str.strip()

print("\nParent Education Level:")
print(df['Demo_ParentEducation'].value_counts())
print(f"Missing: {df['Demo_ParentEducation'].isna().sum()}")
```

```python
df['PolicyArea'] = df['PolicyArea'].str.strip()

print("Policy Areas:")
print(df['PolicyArea'].value_counts())
print(f"\nMissing: {df['PolicyArea'].isna().sum()}")
```

```python
civic_likert_cols = [
    'CivicFair_PolicyConnection',
    'CivicFair_CampusUnderstanding',
    'CivicFair_UsedInBreakout',
    'CivicFair_PreparedForBreakout'
]

# Clean whitespace
for col in civic_likert_cols:
    df[col] = df[col].str.strip()

# Check unique values
print("Civic Fair Likert Scale Values:")
for col in civic_likert_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())

likert_map = {
    'Strongly Disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly Agree': 5
}

for col in civic_likert_cols:
    df[f'{col}_Num'] = df[col].map(likert_map)

print("\nNumeric versions created for Civic Fair questions")
```

```python
# Text/multiple-choice fields
civic_text_cols = [
    'CivicFair_Orgs',
    'CivicFair_OrgsMissed',
    'CivicFair_Learning',
    'CivicFair_Learning_Other'
]

for col in civic_text_cols:
    df[col] = df[col].str.strip()

# Check content
print("\nCivic Fair Text Fields - Sample:")
for col in civic_text_cols:
    print(f"\n{col}:")
    samples = df[col].dropna().head(3)
    for idx, val in samples.items():
        print(f"  - {val}")
```

```python
breakout_likert_cols = [
    'Breakout_POLSPreparation',
    'Breakout_Participation',
    'Breakout_ConsultantContribution',
    'Breakout_DiscussionBalance',
    'Breakout_CivicFairConnection'
]

for col in breakout_likert_cols:
    df[col] = df[col].str.strip()
    df[f'{col}_Num'] = df[col].map(likert_map)

print("Breakout Likert columns cleaned and converted")
```

```python
df['Breakout_Learning'] = df['Breakout_Learning'].str.strip()
df['Breakout_Learning_Other'] = df['Breakout_Learning_Other'].str.strip()

print("\nBreakout Learning - Sample:")
print(df['Breakout_Learning'].dropna().head(3).tolist())
```

```python
action_likert_cols = [
    'Action_TalkToExperts',
    'Action_ConsultantKnowledge',
    'Action_CivicFairConnection',
    'Action_BreakoutConnection'
]

for col in action_likert_cols:
    df[col] = df[col].str.strip()
    df[f'{col}_Num'] = df[col].map(likert_map)

action_text_cols = [
    'Action_StationsChosen',
    'Action_StationsWanted',
    'Action_Learning',
    'Action_Learning_Other'
]

for col in action_text_cols:
    df[col] = df[col].str.strip()

print("\nAction Stations Chosen - Sample:")
print(df['Action_StationsChosen'].dropna().head(3).tolist())
```

```python
overall_cols = [
    'Overall_Connection_Before', 'Overall_Connection_After',
    'Overall_AcademicConf_Before', 'Overall_AcademicConf_After',
    'Overall_ChangeConf_Before', 'Overall_ChangeConf_After'
]

for col in overall_cols:
    df[col] = df[col].str.strip()
    df[f'{col}_Num'] = df[col].map(likert_map)

print("Overall Impact columns cleaned")

# Check values
for col in overall_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

```python
# Recommendation (likely Yes/No)
df['Recommend'] = df['Recommend'].str.strip()
print("Recommendation:")
print(df['Recommend'].value_counts())

# Overall feedback (open text)
df['Overall_Feedback'] = df['Overall_Feedback'].str.strip()

feedback_count = df['Overall_Feedback'].notna().sum()
print(f"\nFeedback responses: {feedback_count} ({feedback_count/len(df)*100:.1f}%)")

print("\nSample feedback:")
for val in df['Overall_Feedback'].dropna().head(2):
    print(f"  - {val[:200]}...")  # First 200 chars
```

```python
# These columns have multiple selections
multi_choice_cols = {
    'CivicFair_Learning': 'CFL',
    'Breakout_Learning': 'BL',
    'Action_StationsChosen': 'ASC',
    'Action_StationsWanted': 'ASW',
    'Action_Learning': 'AL'
}

def create_binary_columns(df, column_name, prefix):
    """Create binary True/False columns from comma-separated values"""
    
    # Get all unique options
    all_options = set()
    for val in df[column_name].dropna():
        options = [x.strip() for x in str(val).split(',')]
        all_options.update(options)
    
    print(f"\n{column_name} → {len(all_options)} unique options found")
    for i, option in enumerate(sorted(all_options), 1):
        print(f"  {i}. {option}")
    
    # Create binary column for each option
    for option in sorted(all_options):
        # Clean name for column
        clean_name = option[:60].replace(' ', '_').replace(',', '')
        clean_name = clean_name.replace('.', '').replace('?', '').replace('!', '')
        clean_name = clean_name.replace("'", '').replace('"', '')
        
        col_name = f'{prefix}_{clean_name}'
        df[col_name] = df[column_name].fillna('').str.contains(option, regex=False)
    
    return df

# Apply to all
print("="*80)
print("CREATING BINARY COLUMNS")
print("="*80)

for col, prefix in multi_choice_cols.items():
    df = create_binary_columns(df, col, prefix)

print(f"\nNew dataframe shape: {df.shape}")
```

```python
# Find duplicates
duplicates = df[df.duplicated(subset=['StudentID'], keep=False)]

print(f"Duplicate StudentIDs: {len(duplicates)}")

if len(duplicates) > 0:
    print("\nDuplicates found:")
    print(duplicates[['StudentID', 'StartDate', 'Duration']].sort_values('StudentID'))
    
    # Keep first submission
    print(f"\nBefore removing duplicates: {len(df)}")
    df = df.drop_duplicates(subset=['StudentID'], keep='first')
    print(f"After removing duplicates: {len(df)}")
else:
    print("No duplicates found!")
```

```python
# 1. Quality flag based on duration
df['Quality_Flag'] = 'Good'
df.loc[df['Duration'] < 300, 'Quality_Flag'] = 'Rushed'  # < 5 minutes

# 2. Count missing substantive questions
substantive_cols = (
    civic_likert_cols + breakout_likert_cols + 
    action_likert_cols + overall_cols +
    ['PolicyArea', 'Recommend']
)

df['Missing_Count'] = df[substantive_cols].isna().sum(axis=1)

# 3. Flag high missing
df.loc[df['Missing_Count'] > 10, 'Quality_Flag'] = 'High_Missing'

print("\nQuality Flag Distribution:")
print(df['Quality_Flag'].value_counts())

print("\nMissing Count Statistics:")
print(df['Missing_Count'].describe())
```

```python
# Create detailed missing data report
missing_summary = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isna().sum(),
    'Missing_Percent': (df.isna().sum() / len(df) * 100).round(2)
})

# Filter to show only columns with missing data
missing_summary = missing_summary[
    missing_summary['Missing_Count'] > 0
].sort_values('Missing_Count', ascending=False)

print("\n" + "="*80)
print("MISSING DATA SUMMARY")
print("="*80)
print(missing_summary.to_string(index=False))

# Save to file
missing_summary.to_csv('Spring2024_Missing_Data_Report.csv', index=False)
print("\nMissing data report saved to file")
```

```python
# Logical order
final_cols = [
    # Identifiers & Metadata
    'StudentID', 'StartDate', 'EndDate', 'RecordedDate',
    'Duration', 'Progress', 'Finished', 'Status',
    'LocationLatitude', 'LocationLongitude',
    
    # Demographics
    'Demo_HispanicLatinx', 'Demo_Race', 'Demo_Gender', 'Demo_ParentEducation',
    
    # Policy
    'PolicyArea',
    
    # Civic Engagement Fair
    'CivicFair_Orgs', 'CivicFair_OrgsMissed',
    'CivicFair_Learning', 'CivicFair_Learning_Other',
    'CivicFair_PolicyConnection', 'CivicFair_CampusUnderstanding',
    'CivicFair_UsedInBreakout', 'CivicFair_PreparedForBreakout',
    
    # Breakout Rooms
    'Breakout_Learning', 'Breakout_Learning_Other',
    'Breakout_POLSPreparation', 'Breakout_Participation',
    'Breakout_ConsultantContribution', 'Breakout_DiscussionBalance',
    'Breakout_CivicFairConnection',
    
    # Action Symposium
    'Action_StationsChosen', 'Action_StationsWanted',
    'Action_Learning', 'Action_Learning_Other',
    'Action_TalkToExperts', 'Action_ConsultantKnowledge',
    'Action_CivicFairConnection', 'Action_BreakoutConnection',
    
    # Overall Impact
    'Overall_Connection_Before', 'Overall_Connection_After',
    'Overall_AcademicConf_Before', 'Overall_AcademicConf_After',
    'Overall_ChangeConf_Before', 'Overall_ChangeConf_After',
    
    # Recommendation & Feedback
    'Recommend', 'Overall_Feedback',
    
    # Quality
    'Quality_Flag', 'Missing_Count'
]

# Add all numeric and binary columns
numeric_cols = [col for col in df.columns if col.endswith('_Num')]
binary_cols = [col for col in df.columns if col.startswith(('CFL_', 'BL_', 'ASC_', 'ASW_', 'AL_'))]

final_cols.extend(numeric_cols)
final_cols.extend(binary_cols)

# Reorder
existing_cols = [col for col in final_cols if col in df.columns]
df = df[existing_cols]

print(f"\nFinal column count: {len(df.columns)}")
```

```python
# Save
df.to_csv('THM_Spring_2024_CLEANED.csv', index=False)

print("\n" + "="*80)
print("CLEANING COMPLETE!")
print("="*80)
print(f"Original rows: 91")
print(f"Final rows: {len(df)}")
print(f"Rows removed: {91 - len(df)}")
print(f"Final columns: {len(df.columns)}")
print(f"\nFile saved: THM_Spring_2024_CLEANED.csv")
```
