```python
# ── Cell 2: Authenticate + full permission check ──────────────────────
#
# This will:
#   1. Open a browser window for you to log in with your Google account
#   2. Verify READ access to the main data folder
#   3. Verify WRITE access to the outputs folder
#   4. Confirm every data file ID in config.py is accessible

from utils.gdrive_auth import DriveDataLoader
from config import MAIN_FOLDER_ID, OUTPUT_FOLDER_ID, GDRIVE_FILES

loader = DriveDataLoader()

all_ok = loader.run_full_check(
    main_folder_id   = MAIN_FOLDER_ID,
    output_folder_id = OUTPUT_FOLDER_ID,
    gdrive_files     = GDRIVE_FILES
)

if not all_ok:
    raise SystemExit('\n⛔ Access check failed. Fix the issues above before proceeding.')
```
